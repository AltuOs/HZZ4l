```c++
#include<TFile.h>
#include<TH1.h>
#include<TMath.h>
#include<math.h>
#include<TF1.h>
#include<TLine.h>
#include<TPoint.h>
#include<iostream>
#include<string>
#include<TLatex.h>
#include<stdio.h>
```

```c++
TCanvas *c1 = new TCanvas("c1","c1",900, 600);
c1->SetGrid();
```

```c++
// Breit-Wigner probability density function: https://root.cern.ch/root/html524/TMath.html#TMath:BreitWigner
TF1 *signal1 = new TF1("signal1","TMath::BreitWigner(x,[0],[1])",-5,5);
signal1->SetName("");
signal1->FixParameter(0,0);
signal1->FixParameter(1,1);
signal1->SetLineColor(kRed);
```

```c++
TF1 *signal2 = new TF1("signal2","TMath::BreitWigner(x,[0],[1])",-5,5);
signal2->SetName("");
signal2->FixParameter(0,0);
signal2->FixParameter(1,2);
signal2->SetLineColor(kGreen);
```

```c++
TF1 *signal3 = new TF1("signal3","TMath::BreitWigner(x,[0],[1])",-5,5);
signal3->SetName("");
signal3->FixParameter(0,2);
signal3->FixParameter(1,0.7);
signal3->SetLineColor(kBlue);
```

```c++
// Create legend
auto legend = new TLegend(0.7,0.7,0.9,0.9);
legend->AddEntry(signal1,"x_{0}=0,  #Gamma=1","L");
legend->AddEntry(signal2,"x_{0}=0,  #Gamma=2","L");
legend->AddEntry(signal3,"x_{0}=2, #Gamma=0.7","L");
```

```c++
// Integral bajo la curva
signal1->Integral(-5,5)
```

```c++
signal1->SetTitle("");
signal1->GetXaxis()->SetTitle("x");
signal1->GetYaxis()->SetTitle("Relative probability (a.u.)");
signal1->GetYaxis()->SetRangeUser(0,1);
signal1->Draw();
signal2->Draw("SAME");
signal3->Draw("SAME");
legend->Draw("SAME");
c1->Draw();

// Save graph
c1->SaveAs("BreitWigner_distribution.png");
```
